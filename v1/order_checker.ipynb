```python
import pandas as pd
import ujson
```

```python
order = pd.read_csv('order.csv')
order
```

```python
order['Qty'].sum()
```

```python
with open("purchase_list.json") as json_file:
    cards = ujson.load(json_file)
# Handling duplicates in json file
new_data = {}
for key, value in cards.items():
    if key in new_data:
        new_data[key] += value
    else:
        new_data[key] = value
cards = new_data.copy()
cards_df = pd.DataFrame(list(cards.items()), columns=['Item', 'Qty'])
cards_df
```

```python
cards_df['Qty'].sum()
```

```python
trimmed_order = order[['Item', 'Qty']].groupby(['Item']).sum(numeric_only=True).reset_index()
trimmed_order
```

```python
# Merge the two dataframes on the specified key column
merged_df = pd.merge(cards_df, trimmed_order, on='Item')
# Subtract the 'qty' column of df2 from df1
merged_df['qty_difference'] = merged_df['Qty_x'] - merged_df['Qty_y']
# Drop the original 'qty' columns
merged_df = merged_df.drop(columns=['Qty_x', 'Qty_y'])
merged_df
```
